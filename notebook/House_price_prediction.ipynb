```python
# è¨“ç·´æ¨¡å‹

# 1ï¸âƒ£ è¼‰å…¥æ•¸æ“š
df = pd.read_csv("test.csv")  # è¨“ç·´é›†

# 2ï¸âƒ£ é€²è¡Œ Target Encodingï¼ˆä½¿ç”¨å€ & é‡Œ çš„å‡åƒ¹ï¼‰
df["district_mean_price"] = df.groupby("å€å")["ç¸½åƒ¹å…ƒ"].transform("mean")
df["village_mean_price"] = df.groupby("é‡Œå")["ç¸½åƒ¹å…ƒ"].transform("mean")

# 3ï¸âƒ£ æ¨™æº–åŒ–
scaler = StandardScaler()
df[["é‡Œå_target", "å€å_target"]] = scaler.fit_transform(df[["village_mean_price", "district_mean_price"]])

# 4ï¸âƒ£ åˆªé™¤ä¸éœ€è¦çš„æ¬„ä½
df.drop(columns=["å€å", "é‡Œå", "district_mean_price", "village_mean_price"], inplace=True)

# 5ï¸âƒ£ å»ºç«‹äº¤å‰è®Šæ•¸ã€è½‰æ›æˆlogå–®åƒ¹
df["district_village_interaction"] = df["é‡Œå_target"] * df["å€å_target"]
df["log_unit_price"] = np.log1p(df["å–®åƒ¹"])  # Log è®Šæ›å–®åƒ¹
df.drop(columns=["å–®åƒ¹"], inplace=True)

# 6ï¸âƒ£ è¨­å®š X èˆ‡ y
X = df.drop(columns=["ç¸½åƒ¹å…ƒ"])  # ç¨ç«‹è®Šæ•¸
y = df["ç¸½åƒ¹å…ƒ"]  # ç›®æ¨™è®Šæ•¸

# 7ï¸âƒ£ è½‰æ›æˆ XGBoost DMatrix
dtrain = xgb.DMatrix(X, label=y)

# 8ï¸âƒ£ è¨­å®š XGBoost æœ€ä½³åƒæ•¸
params = {
    "objective": "reg:squarederror",
    "eval_metric": "mae",
    "colsample_bytree": 1.0,
    "learning_rate": 0.0185,
    "max_depth": 7,
    "subsample": 0.5,
    "min_child_weight": 1,
    "reg_lambda": 0.0021,
    "reg_alpha": 0
}

# 9ï¸âƒ£ é€²è¡Œ 5-Fold äº¤å‰é©—è­‰ï¼ˆé©—è­‰æœ€çµ‚æ¨¡å‹è¡¨ç¾ï¼‰
cv_results = xgb.cv(params, dtrain, num_boost_round=500, nfold=5, early_stopping_rounds=50, metrics="mae", seed=42)

# ğŸ”Ÿ å–å¾—æœ€å°çš„ MAE
best_mae = cv_results["test-mae-mean"].min()
print(f"\nğŸ¯ äº¤å‰é©—è­‰ MAE: {best_mae:,.0f}")

# 1ï¸âƒ£1ï¸âƒ£ ä½¿ç”¨æ‰€æœ‰æ•¸æ“šè¨“ç·´æœ€çµ‚æ¨¡å‹
num_boost_round = len(cv_results)  # å–å¾—æœ€ä½³è¿­ä»£æ¬¡æ•¸
final_model = xgb.train(params, dtrain, num_boost_round=num_boost_round)

# 1ï¸âƒ£2ï¸âƒ£ å„²å­˜æ¨¡å‹
final_model.save_model("xgboost_final_model.json")
print("\nâœ… æœ€çµ‚æ¨¡å‹å·²å„²å­˜ç‚º xgboost_final_model.json")
```

```python
# æ¨¡å‹API

# å®šç¾©å‡½æ•¸ï¼šæª¢æŸ¥æ˜¯å¦ç‚ºæ•¸å€¼
def is_numeric(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

# å®šç¾©å‡½æ•¸ï¼šæª¢æŸ¥æ˜¯å¦ç‚ºäºŒå…ƒæ•¸ (0 æˆ– 1)
def is_binary(value):
    return value in [0, 1]

# å®šç¾©å‡½æ•¸ï¼šæª¢æŸ¥æ˜¯å¦ç‚ºæŒ‡å®šç¯„åœå…§çš„æ•´æ•¸
def is_in_range(value, min_val, max_val):
    try:
        return min_val <= int(value) <= max_val
    except ValueError:
        return False

# å®šç¾©å‡½æ•¸ï¼šè‡ªå‹•å¡«å…¥å€å
def get_district_name(district_code):
    district_mapping = {
        1: "East District",
        2: "North District",
        3: "Xiangshan District"
    }
    return district_mapping.get(district_code, None)

# å®šç¾©å‡½æ•¸ï¼šOne-Hot Encoding
def get_one_hot(value, categories):
    return [1 if value == cat else 0 for cat in categories]

# å®šç¾©å‡½æ•¸ï¼šTarget Encoding
def apply_target_encoding(value, column, target, train_data):
    median_encoding = train_data.groupby(column)[target].median()
    default_value = median_encoding.median()  # è¨­å®šé è¨­å€¼
    return median_encoding.get(value, default_value)

# è®€å–è¨“ç·´è³‡æ–™ï¼Œç”¨æ–¼ Target Encoding
train_data = pd.read_csv("test.csv")

# è¼¸å…¥æˆ¿å±‹è³‡è¨Š
data = {}

# 1. åœŸåœ°ç§»è½‰ç¸½é¢ç©å¹³æ–¹å…¬å°º
while True:
    value = input("è«‹è¼¸å…¥åœŸåœ°ç§»è½‰ç¸½é¢ç©å¹³æ–¹å…¬å°ºï¼š")
    if is_numeric(value):
        data["åœŸåœ°ç§»è½‰ç¸½é¢ç©å¹³æ–¹å…¬å°º"] = float(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•¸å€¼ï¼")

# 2. äº¤æ˜“å¹´æœˆ
while True:
    value = input("è«‹è¼¸å…¥äº¤æ˜“å¹´æœˆï¼ˆç¯„ä¾‹ï¼š11403 è¡¨ç¤ºæ°‘åœ‹114å¹´3æœˆï¼‰ï¼š")
    if value.isdigit() and len(value) == 5:
        data["äº¤æ˜“å¹´æœˆ"] = int(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥5ä½æ•¸çš„æ°‘åœ‹å¹´æœˆä»½ï¼ˆä¾‹å¦‚11403ï¼‰ï¼")

# 3. å»ºç‰©æ˜¯å¦å¤§æ–¼1
while True:
    value = input("å»ºç‰©æ˜¯å¦å¤§æ–¼1 (0=å¦, 1=æ˜¯)ï¼š")
    if is_binary(int(value)):
        data["å»ºç‰©å¤§æ–¼1"] = int(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥0æˆ–1ï¼")

# 4. è»Šä½æ•¸
while True:
    value = input("è«‹è¼¸å…¥è»Šä½æ•¸ï¼š")
    if value.isdigit():
        data["è»Šä½æ•¸"] = int(value)
        if data["è»Šä½æ•¸"] == 0:
            data["è»Šä½ç¸½åƒ¹å…ƒ"] = 0  # è‹¥è»Šä½æ•¸ç‚º0ï¼Œè»Šä½ç¸½åƒ¹å…ƒè‡ªå‹•å¡«0
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•´æ•¸ï¼")

# 5. æ¨“å±¤æ¯”ä¾‹
while True:
    value = input("è«‹è¼¸å…¥æ¨“å±¤æ¯”ä¾‹ (0~1ï¼Œä¾‹å¦‚10/15=0.67)ï¼š")
    if is_numeric(value) and 0 <= float(value) <= 1:
        data["æ¨“å±¤æ¯”ä¾‹"] = float(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥0åˆ°1ä¹‹é–“çš„æ•¸å€¼ï¼")

# 6. æ˜¯å¦ç‚ºé€å¤©
while True:
    value = input("æ˜¯å¦ç‚ºé€å¤© (0=å¦, 1=æ˜¯)ï¼š")
    if is_binary(int(value)):
        data["æ˜¯å¦é€å¤©"] = int(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥0æˆ–1ï¼")

# 7. æ˜¯å¦ç‚ºæŠ•è³‡é€å¤©
while True:
    value = input("æ˜¯å¦ç‚ºæŠ•è³‡é€å¤© (0=å¦, 1=æ˜¯)ï¼š")
    if is_binary(int(value)):
        data["æ˜¯å¦æŠ•è³‡é€å¤©"] = int(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥0æˆ–1ï¼")

# 8. å±‹é½¡æ˜¯å¦ç¼ºå¤±
while True:
    value = input("å±‹é½¡æ˜¯å¦ç¼ºå¤± (0=å¦, 1=æ˜¯)ï¼š")
    if is_binary(int(value)):
        data["å±‹é½¡ç¼ºå¤±"] = int(value)
        if data["å±‹é½¡ç¼ºå¤±"] == 1:
            data["å±‹é½¡"] = 0  # è‹¥å±‹é½¡ç¼ºå¤±ç‚º0ï¼Œå±‹é½¡è‡ªå‹•å¡«0
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥0æˆ–1ï¼")

# 9. å±‹é½¡
if data["å±‹é½¡ç¼ºå¤±"] == 0:
    while True:
        value = input("è«‹è¼¸å…¥å±‹é½¡ï¼š")
        if is_numeric(value):
            data["å±‹é½¡"] = float(value)
            break
        else:
            print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•¸å€¼ï¼")

# 10. å»ºç‰©ç§»è½‰ç¸½é¢ç©å¹³æ–¹å…¬å°º
while True:
    value = input("è«‹è¼¸å…¥å»ºç‰©ç§»è½‰ç¸½é¢ç©å¹³æ–¹å…¬å°ºï¼š")
    if is_numeric(value):
        data["å»ºç‰©ç§»è½‰ç¸½é¢ç©å¹³æ–¹å…¬å°º"] = float(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•¸å€¼ï¼")

# 11. æ˜¯å¦æ¯›èƒš
while True:
    value = input("æ˜¯å¦æ¯›èƒš (0=å¦, 1=æ˜¯)ï¼š")
    if is_binary(int(value)):
        data["æ˜¯å¦æ¯›èƒš"] = int(value)
        if data["æ˜¯å¦æ¯›èƒš"] == 1:
            data["å»ºç‰©ç¾æ³æ ¼å±€-æˆ¿"] = 0  # è‹¥æ˜¯å¦æ¯›èƒšç‚º0ï¼Œå»ºç‰©æˆ¿é–“æ•¸è‡ªå‹•å¡«0
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥0æˆ–1ï¼")

# 12. å»ºç‰©æˆ¿é–“æ•¸
if data["æ˜¯å¦æ¯›èƒš"] == 0:
    while True:
        value = input("è«‹è¼¸å…¥å»ºç‰©æˆ¿é–“æ•¸ï¼š")
        if value.isdigit():
            data["å»ºç‰©ç¾æ³æ ¼å±€-æˆ¿"] = int(value)
            break
        else:
            print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•´æ•¸ï¼")

# 13. æ˜¯å¦æœ‰ç®¡ç†çµ„ç¹”
while True:
    value = input("æ˜¯å¦æœ‰ç®¡ç†çµ„ç¹” (0=å¦, 1=æ˜¯)ï¼š")
    if is_binary(int(value)):
        data["æœ‰ç„¡ç®¡ç†çµ„ç¹”"] = int(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥0æˆ–1ï¼")

# 14. è»Šä½ç¸½åƒ¹å…ƒ
if data["è»Šä½æ•¸"] > 0:
    while True:
        value = input("è«‹è¼¸å…¥è»Šä½ç¸½åƒ¹å…ƒï¼š")
        if is_numeric(value):
            data["è»Šä½ç¸½åƒ¹å…ƒ"] = float(value)
            break
        else:
            print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•¸å€¼ï¼")

# 15. æ˜¯å¦ç‚ºç‰¹æ®Šäº¤æ˜“
while True:
    value = input("æ˜¯å¦ç‚ºç‰¹æ®Šäº¤æ˜“ (0=å¦, 1=æ˜¯)ï¼š")
    if is_binary(int(value)):
        data["ç‰¹æ®Šäº¤æ˜“"] = int(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥0æˆ–1ï¼")

# 16. æ˜¯å¦æœ‰é›»æ¢¯
while True:
    value = input("æ˜¯å¦æœ‰é›»æ¢¯ (0=å¦, 1=æ˜¯)ï¼š")
    if is_binary(int(value)):
        data["é›»æ¢¯"] = int(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥0æˆ–1ï¼")

# 17. ç·¯åº¦
while True:
    value = input("è«‹è¼¸å…¥ç·¯åº¦ï¼ˆå‚™è¨»ï¼šGoogleæŸ¥è©¢åœ°å€æ™‚ï¼Œç¶²å€æ¬„æœƒé¡¯ç¤ºï¼‰ï¼š")
    if is_numeric(value):
        data["latitude"] = float(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•¸å€¼ï¼")

# 18. ç¶“åº¦
while True:
    value = input("è«‹è¼¸å…¥ç¶“åº¦ï¼ˆå‚™è¨»ï¼šGoogleæŸ¥è©¢åœ°å€æ™‚ï¼Œç¶²å€æ¬„æœƒé¡¯ç¤ºï¼‰ï¼š")
    if is_numeric(value):
        data["longitude"] = float(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•¸å€¼ï¼")

# 19. èˆ‡ç«¹ç§‘çš„è·é›¢
while True:
    value = input("è«‹è¼¸å…¥èˆ‡ç«¹ç§‘çš„è·é›¢ (km)ï¼š")
    if is_numeric(value):
        data["èˆ‡ç«¹ç§‘è·é›¢_km"] = float(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥æ•¸å€¼ï¼")

# 20. æ˜¯å¦ç‚ºé«˜è£ä¿®æˆ¿
while True:
    value = input("æ˜¯å¦ç‚ºé«˜è£ä¿®æˆ¿ (0=å¦, 1=æ˜¯)ï¼š")
    if is_binary(int(value)):
        data["high_price_decorated"] = int(value)
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥0æˆ–1ï¼")

# 21. éƒ½å¸‚åœŸåœ°ä½¿ç”¨åˆ†å€
zone_options = ["ä½", "å…¶", "å•†", "å·¥", "è¾²", "é"]
while True:
    value = input("è«‹è¼¸å…¥éƒ½å¸‚åœŸåœ°ä½¿ç”¨åˆ†å€ï¼š(1)ä½ (2)å…¶ (3)å•† (4)å·¥ (5)è¾² (6)éï¼š")
    if is_in_range(value, 1, 6):
        data.update(dict(zip([f"éƒ½å¸‚åœŸåœ°ä½¿ç”¨åˆ†å€_{z}" for z in zone_options], get_one_hot(zone_options[int(value) - 1], zone_options))))
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥1åˆ°6ä¹‹é–“çš„æ•´æ•¸ï¼")

# 22. å»ºç‰©å‹æ…‹
building_types = ["å…¬å¯“", "å¤§æ¨“", "è¯å»ˆ", "é€å¤©å"]
while True:
    value = input("è«‹è¼¸å…¥å»ºç‰©å‹æ…‹ï¼š(1)å…¬å¯“ (2)å¤§æ¨“ (3)è¯å»ˆ (4)é€å¤©åï¼š")
    if is_in_range(value, 1, 4):
        data.update(dict(zip([f"å»ºç‰©å‹æ…‹_{b}" for b in building_types], get_one_hot(building_types[int(value) - 1], building_types))))
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥1åˆ°4ä¹‹é–“çš„æ•´æ•¸ï¼")

# 23. ä¸»è¦ç”¨é€”
usage_types = ["ä½å•†ç”¨", "ä½å®¶ç”¨", "å…¶ä»–", "å•†æ¥­ç”¨", "å•†è¾¦ç”¨", "å·¥æ¥­ç”¨", "è¾¦å…¬ç”¨", "è¾²æ¥­ç”¨"]
while True:
    value = input("è«‹è¼¸å…¥ä¸»è¦ç”¨é€”ï¼š(1)ä½å•†ç”¨ (2)ä½å®¶ç”¨ (3)å…¶ä»– (4)å•†æ¥­ç”¨ (5)å•†è¾¦ç”¨ (6)å·¥æ¥­ç”¨ (7)è¾¦å…¬ç”¨ (8)è¾²æ¥­ç”¨ï¼š")
    if is_in_range(value, 1, 8):
        data.update(dict(zip([f"ä¸»è¦ç”¨é€”_{u}" for u in usage_types], get_one_hot(usage_types[int(value) - 1], usage_types))))
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥1åˆ°8ä¹‹é–“çš„æ•´æ•¸ï¼")

# 24. è»Šä½é¡åˆ¥
parking_types = ["å…¶ä»–", "å‡é™å¹³é¢", "å‡é™æ©Ÿæ¢°", "å¡é“æ©Ÿæ¢°", "å¹³é¢", "ç„¡"]
while True:
    value = input("è«‹è¼¸å…¥è»Šä½é¡åˆ¥ï¼š(1)å…¶ä»– (2)å‡é™å¹³é¢ (3)å‡é™æ©Ÿæ¢° (4)å¡é“æ©Ÿæ¢° (5)å¹³é¢ (6)ç„¡ï¼š")
    if is_in_range(value, 1, 6):
        data.update(dict(zip([f"è»Šä½é¡åˆ¥_{p}" for p in parking_types], get_one_hot(parking_types[int(value) - 1], parking_types))))
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥1åˆ°6ä¹‹é–“çš„æ•´æ•¸ï¼")

# 25. å€å
while True:
    value = input("è«‹è¼¸å…¥å€åï¼ˆ1=æ±å€, 2=åŒ—å€, 3=é¦™å±±å€ï¼‰ï¼š")
    if is_in_range(value, 1, 3):
        data["å€å"] = get_district_name(int(value))
        break
    else:
        print("âš ï¸ è¼¸å…¥éŒ¯èª¤ï¼šè«‹è¼¸å…¥1åˆ°3ä¹‹é–“çš„æ•´æ•¸ï¼")

# 26. é‡Œå
data["é‡Œå"] = input("è«‹è¼¸å…¥é‡Œåï¼š")

# è¨ˆç®—å€åå’Œé‡Œåçš„ Target Encoding
data["å€å_target"] = apply_target_encoding(data["å€å"], "å€å", "mean_price_per_unit", train_data)
data["é‡Œå_target"] = apply_target_encoding(data["é‡Œå"], "é‡Œå", "mean_price_per_unit", train_data)

# å°‡è¼¸å…¥çš„è³‡æ–™è½‰æ›ç‚º DataFrame
input_df = pd.DataFrame([data])

# è¨ˆç®— district_village_interaction
input_df["district_village_interaction"] = input_df["å€å_target"] * input_df["é‡Œå_target"]

# è¼‰å…¥å–®åƒ¹æŸ¥è©¢è¡¨
price_lookup_table = pd.read_csv("price_lookup_table.csv")

# å°‡æŸ¥è©¢è¡¨èˆ‡è¼¸å…¥è³‡æ–™åˆä½µ
input_df = pd.merge(input_df, price_lookup_table, on=["å€å", "é‡Œå"], how="left")

# å¦‚æœæŸäº›å€åŸŸæ²’æœ‰å°æ‡‰çš„å¹³å‡å–®åƒ¹ï¼Œä½¿ç”¨å…¨å±€å¹³å‡å–®åƒ¹å¡«å……
global_average_unit_price = price_lookup_table["average_unit_price"].mean()
input_df["average_unit_price"].fillna(global_average_unit_price, inplace=True)

# è¨ˆç®— log_unit_price
input_df["log_unit_price"] = np.log1p(input_df["average_unit_price"])

# è¨ˆç®—é æ¸¬ç¸½åƒ¹
input_df["predicted_total_price"] = input_df["average_unit_price"] * input_df["å»ºç‰©ç§»è½‰ç¸½é¢ç©å¹³æ–¹å…¬å°º"]

# è¼¸å‡ºæœ€çµ‚é æ¸¬çµæœ
print("\nğŸ¯ é æ¸¬çµæœï¼š")
print(f"å¹³å‡å–®åƒ¹: {input_df['average_unit_price'].values[0]:,.2f} å…ƒ/å¹³æ–¹å…¬å°º")
print(f"é æ¸¬ç¸½åƒ¹: {input_df['predicted_total_price'].values[0]:,.0f} å…ƒ")
```
